# R^2 Results 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load all CSV files
df_linear = pd.read_csv("results_linear_reg.csv")
df_tree = pd.read_csv("results_decision_tree.csv")
df_boost = pd.read_csv("results_simple_tree_boosting.csv")
df_forest = pd.read_csv("results_random_forest.csv")

# Constants
num_homes = 3913
num_rooms = 1476
sst_homes = 31637397.457705088
sst_rooms = 5951703.530487805
total_sst = sst_homes + sst_rooms

# --- Linear Regression ---
pivot_df = df_linear.pivot(index="Iteration", columns="Dataset", values="Validation_MSE")
pivot_df['Weighted_MSE'] = (
    pivot_df['Homes'] * num_homes + pivot_df['Private Rooms'] * num_rooms
)
pivot_df['R2_Score'] = 1 - (pivot_df['Weighted_MSE'] / total_sst)

best_lr_iter = pivot_df['R2_Score'].idxmax()
best_lr_r2 = pivot_df.loc[best_lr_iter, 'R2_Score']
best_lr_cv = pivot_df.loc[best_lr_iter, 'Weighted_MSE']

# --- Decision Tree ---
df_tree_sorted = df_tree.sort_values('Iteration')
best_tree_row = df_tree_sorted.loc[df_tree_sorted['R_Squared'].idxmax()]
best_tree_iter = int(best_tree_row['Iteration'])
best_tree_r2 = best_tree_row['R_Squared']
best_tree_cv = best_tree_row['Validation_MSE']

# --- Boosted Tree ---
df_boost_sorted = df_boost.sort_values('Iteration')
best_boost_row = df_boost_sorted.loc[df_boost_sorted['R_Squared'].idxmax()]
best_boost_iter = int(best_boost_row['Iteration'])
best_boost_r2 = best_boost_row['R_Squared']
best_boost_cv = best_boost_row['Validation_MSE']

# --- Random Forest ---
df_forest_sorted = df_forest.sort_values('Iteration')
best_forest_row = df_forest_sorted.loc[df_forest_sorted['R_Squared'].idxmax()]
best_forest_iter = int(best_forest_row['Iteration'])
best_forest_r2 = best_forest_row['R_Squared']
best_forest_cv = best_forest_row['Validation_MSE']

# --- Plotting ---
plt.figure(figsize=(10, 6))

# Linear Regression
plt.plot(pivot_df.index, pivot_df['R2_Score'], marker='o', label="Linear Regression R²")
plt.scatter(best_lr_iter, best_lr_r2)
# Linear Regression (with arrow)
plt.annotate(
    f"Iter {best_lr_iter}\nR²={best_lr_r2:.4f}\nCV={best_lr_cv:.2f}",
    xy=(best_lr_iter, best_lr_r2),
    xytext=(best_lr_iter, best_lr_r2 + 0.01),
    textcoords='data',
    arrowprops=dict(arrowstyle="->"),
    ha='left',
    fontsize=9
)
# Decision Tree
plt.plot(df_tree_sorted['Iteration'], df_tree_sorted['R_Squared'], marker='s', label="Decision Tree R²")
plt.scatter(best_tree_iter, best_tree_r2)
# Decision Tree (with arrow)
plt.annotate(
    f"Iter {best_tree_iter}\nR²={best_tree_r2:.4f}\nCV={best_tree_cv:.2f}",
    xy=(best_tree_iter, best_tree_r2),
    xytext=(best_tree_iter, best_tree_r2 + 0.01),
    textcoords='data',
    arrowprops=dict(arrowstyle="->"),
    ha='left',
    fontsize=9
)
# Boosted Tree
plt.plot(df_boost_sorted['Iteration'], df_boost_sorted['R_Squared'], marker='^', label="Boosted Tree R²")
plt.scatter(best_boost_iter, best_boost_r2)
plt.annotate(
    f"Iter {best_boost_iter}\nR²={best_boost_r2:.4f}\nCV={best_boost_cv:.2f}",
    xy=(best_boost_iter, best_boost_r2),
    xytext=(best_boost_iter, best_boost_r2 - 0.05),
    textcoords='data',
    arrowprops=dict(arrowstyle="->"),
    ha='center',
    fontsize=9
)

# Random Forest
plt.plot(df_forest_sorted['Iteration'], df_forest_sorted['R_Squared'], marker='d', label="Random Forest R²")
plt.scatter(best_forest_iter, best_forest_r2)
plt.annotate(
    f"Iter {best_forest_iter}\nR²={best_forest_r2:.4f}\nCV={best_forest_cv:.2f}",
    xy=(best_forest_iter, best_forest_r2),
    xytext=(best_forest_iter, best_forest_r2 + 0.023),
    textcoords='data',
    arrowprops=dict(arrowstyle="->"),
    ha='center',
    fontsize=9
)

# Final touches
plt.title("R² Score vs Number of Features")
plt.xlabel("Number of Features")
plt.ylabel("R² Score")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
